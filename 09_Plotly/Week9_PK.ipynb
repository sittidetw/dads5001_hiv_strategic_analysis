```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', delimiter='\t')
```

```python
df.head()
```

```python
# แปลงราคาจาก string เป็น float
df['item_price'] = df['item_price'].str.replace('$', '', regex=False).astype(float)
```

```python
import plotly.express as px

# รวมจำนวนชิ้นตามชื่อเมนู
top_items = df.groupby('item_name')['quantity'].sum().reset_index()
top_items = top_items.sort_values(by='quantity', ascending=False).head(5)

# สร้างกราฟแท่ง
fig1 = px.bar(top_items, x='item_name', y='quantity',
             title='Top 5 Most Popular Items (by Quantity)',
             color='quantity',
             text_auto=True)
fig1.show()
```

```python
# คำนวณยอดรวมต่อ 1 order_id
order_total = df.groupby('order_id')['item_price'].sum().reset_index()

# สร้างกราฟ Histogram
fig2 = px.histogram(order_total, x='item_price',
                   nbins=30,
                   title='Distribution of Order Totals',
                   labels={'item_price': 'Order Total ($)'},
                   color_discrete_sequence=['indianred'])
fig2.show()
```

```python
# สร้างคอลัมน์ใหม่เพื่อเช็คว่าเป็นเมนูไก่หรือไม่
df['is_chicken'] = df['item_name'].apply(lambda x: 'Chicken' if 'Chicken' in x else 'Other')

# รวมยอดขาย
chicken_vs_others = df.groupby('is_chicken')['item_price'].sum().reset_index()

# สร้างกราฟวงกลม
fig3 = px.pie(chicken_vs_others, values='item_price', names='is_chicken',
             title='Revenue Proportion: Chicken Items vs Others',
             hole=0.4, # ทำเป็นกราฟ Donut
             color_discrete_map={'Chicken':'#f39c12', 'Other':'#bdc3c7'})
fig3.show()
```

```python
import plotly.express as px

# 1. รันโค้ด Clean ข้อมูลราคาก่อน (ถ้ายังไม่ได้ทำ)
# df['item_price'] = df['item_price'].str.replace('$', '', regex=False).astype(float)

# 2. สร้าง 2D Histogram / Density Heatmap
# แกน X: ราคา, แกน Y: จำนวน
fig = px.density_heatmap(df,
                        x='item_price',
                        y='quantity',
                        nbinsx=20, # จำนวนถัง (Bin) ของแกน X
                        nbinsy=10, # จำนวนถัง (Bin) ของแกน Y
                        title='2D Histogram: Order Quantity vs Item Price Density',
                        labels={'item_price': 'Item Price ($)', 'quantity': 'Quantity ordered'},
                        color_continuous_scale='Viridis', # สีไล่ระดับ
                        text_auto=True # แสดงตัวเลขจำนวนข้อมูลในช่อง
                       )

# ปรับปรุง layout เพื่อให้เห็นช่องสีเหลี่ยมชัดเจนขึ้น
fig.update_layout(
    xaxis=dict(ticks="outside"),
    yaxis=dict(ticks="outside")
)

fig.show()
```
